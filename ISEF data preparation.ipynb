combine two era5 files. official step for prepartion data step 1). 1.1), combine surface and pressure files. convert from K to C and caclulate the rh. convert grib to nc files


```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

!apt-get update -qq && apt-get install -y cdo > /dev/null
!pip install -q xarray pandas numpy > /dev/null

import xarray as xr
import glob
from pathlib import Path

base_path = '/content/drive/MyDrive/graphcast_project/ERA5'

surface_files = sorted(glob.glob(f"{base_path}/era5-*surface-ISEF.grib"))
print("Found years:", [Path(f).stem.split('-')[1] for f in surface_files])

for surf_file in surface_files:
    year = Path(surf_file).stem.split('-')[1]
    pres_file = f"{base_path}/era5-{year}-pressure-ISEF.grib"

    print(f"\nProcessing {year}...")

    # Convert GRIB → NetCDF
    !cdo -f nc copy "{surf_file}"  /content/temp_surf.nc > /dev/null
    !cdo -f nc copy "{pres_file}"  /content/temp_pres.nc > /dev/null

    # Open with scipy engine (always works)
    ds_surf = xr.open_dataset("/content/temp_surf.nc", engine='scipy')
    ds_pres = xr.open_dataset("/content/temp_pres.nc", engine='scipy')

    # Auto-detect latitude/longitude names
    for ds in [ds_surf, ds_pres]:
        # latitude
        if 'latitude' in ds.dims:
            pass
        elif 'lat' in ds.dims:
            ds = ds.rename({'lat': 'latitude'})
        # longitude
        if 'longitude' in ds.dims:
            pass
        elif 'lon' in ds.dims:
            ds = ds.rename({'lon': 'longitude'})
        # make sure latitude increases northward
        if ds.latitude.values[-1] < ds.latitude.values[0]:
            ds = ds.isel(latitude=slice(None, None, -1))

    # Standard variable names
    rename_map = {'10u':'u10','10v':'v10','2t':'t2m','2d':'d2m'}
    ds_surf = ds_surf.rename({k:v for k,v in rename_map.items() if k in ds_surf})
    ds_pres = ds_pres.rename({k:v for k,v in rename_map.items() if k in ds_pres})

    # Merge
    ds_year = xr.merge([ds_surf, ds_pres], compat='override')

    # Add °C and RH
    if 't2m' in ds_year and 'd2m' in ds_year:
        ds_year['t2m_c'] = ds_year['t2m'] - 273.15
        ds_year['d2m_c'] = ds_year['d2m'] - 273.15
        ds_year['rh'] = 100 * (
            np.exp((17.625*ds_year.d2m_c)/(243.04+ds_year.d2m_c)) /
            np.exp((17.625*ds_year.t2m_c)/(243.04+ds_year.t2m_c))
        )
    if 'msl' in ds_year:
        ds_year['msl_hpa'] = ds_year.msl / 100

    # Save
    output_file = f"{base_path}/era5_california_{year}_full.nc"
    ds_year.to_netcdf(output_file, engine='scipy')

    # Safe printing
    lat_name = 'latitude' if 'latitude' in ds_year.dims else list(ds_year.dims)[0]
    lon_name = 'longitude' if 'longitude' in ds_year.dims else list(ds_year.dims)[1]

    print(f"SAVED → {Path(output_file).name}")
    print(f"   Time steps: {len(ds_year.time)} | Grid: {len(ds_year[lat_name])}×{len(ds_year[lon_name])}")

print("\nALL 4 YEARS SUCCESSFULLY SAVED!")
print("Files created:")
for y in ['2022','2023','2024','2025']:
    print(f"   era5_california_{y}_full.nc")
```

1.6) pivot the hpa 700 and 850 data into one row

```python
# FINAL PERFECT WIDE PIVOT — ALL SURFACE + BOTH 850 & 700 hPa
from google.colab import drive
drive.mount('/content/drive')

import xarray as xr
import pandas as pd
from functools import reduce

input_file  = '/content/drive/MyDrive/graphcast_project/ERA5/era5_california_2022_2025_final.nc'
output_file = '/content/drive/MyDrive/graphcast_project/ERA5/era5_california_2022_2025_WIDE_FULL.nc'

print("Loading dataset...")
ds = xr.open_dataset(input_file, engine='scipy')
print(f"Time: {len(ds.time)}, Grid: {len(ds.lat)}×{len(ds.lon)}, plev: {list(ds.plev.values)} hPa")

# Convert once to DataFrame
print("Converting to DataFrame...")
df = ds.to_dataframe().reset_index()

print(f"Total rows before processing: {len(df):,}")

# === Step 1: Extract surface variables (appear once per time/lat/lon) ===
surface_vars = [v for v in ds.data_vars if 'plev' not in ds[v].dims]
print(f"Surface variables ({len(surface_vars)}): {surface_vars}")

df_surface = df[['time', 'lat', 'lon'] + surface_vars].drop_duplicates(subset=['time', 'lat', 'lon'])

# === Step 2: Pivot pressure-level variables ===
pressure_vars = ['z', 't', 'u', 'v']
print(f"Pressure-level variables: {pressure_vars}")

df_list = [df_surface]  # start with surface

for var in pressure_vars:
    print(f"Pivoting {var}...")
    df_var = df[['time', 'lat', 'lon', 'plev', var]].copy()
    df_pivot = df_var.pivot_table(
        index=['time', 'lat', 'lon'],
        columns='plev',
        values=var,
        aggfunc='first'  # safe
    ).reset_index()

    # Rename columns: 85000.0 → z_85000
    plev_cols = {p: f"{var}_{int(p)}" for p in df_pivot.columns[3:]}
    df_pivot = df_pivot.rename(columns=plev_cols)

    df_list.append(df_pivot)

# === Step 3: Merge everything ===
print("Merging all parts...")
df_final = reduce(lambda left, right: pd.merge(left, right, on=['time', 'lat', 'lon'], how='inner'), df_list)

print(f"Final shape: {df_final.shape[0]:,} rows × {df_final.shape[1]} columns")

# === Step 4: Back to xarray ===
print("Converting to xarray Dataset...")
ds_wide = df_final.set_index(['time', 'lat', 'lon']).to_xarray()
ds_wide = ds_wide.sortby('time')

# Save
ds_wide.to_netcdf(output_file, engine='scipy')

print(f"\nPERFECT WIDE FILE SAVED!")
print(f"   → {output_file}")
print(f"   → Total variables: {len(ds_wide.data_vars)}")
print(f"   → Surface vars: {surface_vars}")
print(f"   → Pressure vars: z_85000, t_85000, u_85000, v_85000, z_70000, t_70000, ...")

# Quick check: LA on Jan 1 2022
sample = ds_wide.sel(time='2022-01-01T12', lat=34.05, lon=-118.25, method='nearest')
print(f"\nLA Jan 1 2022 sample:")
print(f"   t2m = {sample['t2m'].item():.2f} K → t2m_c should be {sample.get('t2m_c', sample['t2m']-273.15).item():.2f}°C")
print(f"   z_85000 = {sample['z_85000'].item():.1f}")
print(f"   z_70000 = {sample['z_70000'].item():.1f}")
print(f"   msl = {sample['msl'].item():.0f} Pa")
```

```python
from google.colab import drive
drive.mount('/content/drive')
!pip install xarray pandas numpy scipy
import xarray as xr
import pandas as pd
import numpy as np
from scipy.interpolate import griddata
import zipfile
import glob
import os
import gc

# Configuration
AGG_METHOD = 'mean'  # 'mean' or 'max'

# File paths
pollution_dir = '/content/drive/MyDrive/graphcast_project/Pollution/'
output_file = '/content/drive/MyDrive/graphcast_project/Pollution/pollution_la_2022_to_2025_avg.nc'
era5_file = '/content/drive/MyDrive/graphcast_project/Training Data/era5_california_2022_2025_WIDE_FULL.nc'

# === REALISTIC PHYSICAL BOUNDS (EPA, WHO, California wildfires) ===
valid_ranges = {
    'CO':   (0, 60),     # ppm
    'NO2':  (0, 300),    # ppb
    'O3':   (0, 0.5),    # ppm (500 ppb)
    'PM10': (0, 2000),   # µg/m³
    'PM25': (0, 1500),   # µg/m³ ← your updated value (wildfires hit 1000+)
    'SO2':  (0, 1000)    # ppb
}

pollutants = {
    'CO': 'CO_hourly_42101',
    'NO2': 'NO2_hourly_42602',
    'O3': 'O3_hourly_44201',
    'PM10': 'PM10_hourly_81102',
    'PM25': 'PM25_hourly_88101',
    'SO2': 'SO2_hourly_42401'
}

# ERA5 grid and time
era5_grid = {
    'lat': np.arange(28, 50, 0.25),
    'lon': np.arange(-135, -105, 0.25),
    'time': pd.date_range('2022-01-01 00:00:00', '2025-06-30 18:00:00', freq='6h')
}

os.makedirs(pollution_dir, exist_ok=True)

# [Steps 1–10: Your original code unchanged until interpolation...]
# Step 1: Extract and append CSVs for each pollutant
df_list = []
for pollutant, file_prefix in pollutants.items():
    for year in ['2021', '2022', '2023', '2024', '2025']:
        zip_file = os.path.join(pollution_dir, f"{file_prefix}_{year}.zip")
        if not os.path.exists(zip_file):
            print(f"Step 1a: Warning: {zip_file} not found")
            continue
        try:
            print(f"Step 1b: Processing {zip_file}")
            with zipfile.ZipFile(zip_file, 'r') as z:
                csv_names = [f for f in z.namelist() if f.endswith('.csv')]
                if not csv_names:
                    print(f"Step 1c: No CSV file in {zip_file}")
                    continue
                temp_dir = f'/tmp/pollution_temp_{pollutant}_{year}/'
                os.makedirs(temp_dir, exist_ok=True)
                for f in z.namelist():
                    if f.endswith('.csv'):
                        z.extract(f, temp_dir)

                # Load CSV
                csv_path = os.path.join(temp_dir, csv_names[0])
                df = pd.read_csv(csv_path, low_memory=False)
                df['pollutant'] = pollutant
                print(f"Step 1d: Loaded {len(df)} rows for {pollutant} {year}")
                print(f"Step 1e: Columns: {df.columns.tolist()}")
                df_list.append(df)

                # Clean up
                for f in os.listdir(temp_dir):
                    os.remove(os.path.join(temp_dir, f))
                os.rmdir(temp_dir)

                gc.collect()
        except Exception as e:
            print(f"Step 1f: Error processing {zip_file}: {e}")
            continue

if not df_list:
    raise ValueError("Step 2: No valid pollution data found in zip files.")

# Combine DataFrames
df_pollution = pd.concat(df_list, ignore_index=True)
print(f"Step 3: Combined pollution data: {len(df_pollution)} rows")
print(f"Step 3b: Sample data:\n{df_pollution[['Date GMT', 'Time GMT', 'Sample Measurement', 'Latitude', 'Longitude', 'pollutant', 'Units of Measure']].head()}")

# Step 4: Filter for LA region and time range
df_pollution = df_pollution[
    (df_pollution['Latitude'].between(28 , 50)) &
    (df_pollution['Longitude'].between(-135, -105))
]
print(f"Step 4: After filtering for LA region (lat: 28-50, lon: -135 to -105): {len(df_pollution)} rows")

# Step 5: Parse datetime as UTC using Date GMT and Time GMT
try:
    df_pollution['timestamp'] = pd.to_datetime(
        df_pollution['Date GMT'] + ' ' + df_pollution['Time GMT'],
        errors='coerce',
        utc=True
    )
    invalid_times = df_pollution['timestamp'].isna().sum()
    if invalid_times > 0:
        print(f"Step 5a: Warning: {invalid_times} invalid timestamps found")
        print(f"Step 5b: Sample invalid timestamps:\n{df_pollution[df_pollution['timestamp'].isna()][['Date GMT', 'Time GMT']].head()}")
    df_pollution = df_pollution.dropna(subset=['timestamp'])
    # Filter for data from 2021-12-31 18:00:00 UTC onward
    df_pollution = df_pollution[df_pollution['timestamp'] >= '2021-12-31 18:00:00+00:00']
    print(f"Step 5c: After datetime parsing and time filtering: {len(df_pollution)} rows")
    print(f"Step 5d: Sample UTC timestamps: {df_pollution['timestamp'].head().tolist()}")
    # Check for data in the first 6-hour window
    print(f"Step 5e: Timestamps in 2021-12-31 18:00:00 to 2021-12-31 23:59:59 UTC:\n"
          f"{df_pollution[df_pollution['timestamp'].between('2021-12-31 18:00:00+00:00', '2021-12-31 23:59:59+00:00')][['timestamp', 'pollutant']].head()}")
    print(f"Step 5f: Timestamps in 2022-01-01 00:00:00 to 2022-01-01 05:59:59 UTC:\n"
          f"{df_pollution[df_pollution['timestamp'].between('2022-01-01 00:00:00+00:00', '2022-01-01 05:59:59+00:00')][['timestamp', 'pollutant']].head()}")
except Exception as e:
    print(f"Step 5g: Error parsing timestamps: {e}")
    raise

# Step 6: Parse concentration values
df_pollution['concentration'] = pd.to_numeric(df_pollution['Sample Measurement'], errors='coerce')
df_pollution = df_pollution.dropna(subset=['concentration'])
print(f"Step 6: After parsing concentration: {len(df_pollution)} rows")
print(f"Step 6b: Non-NaN concentration count per pollutant:\n{df_pollution.groupby('pollutant')['concentration'].count()}")

# Step 7: Add coordinates
df_pollution['latitude'] = pd.to_numeric(df_pollution['Latitude'], errors='coerce')
df_pollution['longitude'] = pd.to_numeric(df_pollution['Longitude'], errors='coerce')
df_pollution = df_pollution.dropna(subset=['latitude', 'longitude'])
print(f"Step 7: After adding coordinates: {len(df_pollution)} rows")
print(f"Step 7b: Sample coordinates:\n{df_pollution[['latitude', 'longitude']].head()}")
print(f"Step 7c: Coordinate ranges: lat {df_pollution['latitude'].min()} to {df_pollution['latitude'].max()}, lon {df_pollution['longitude'].min()} to {df_pollution['longitude'].max()}")

# [Imports and Configuration - Keep as is]
# ... (Run your Steps 1 through 7 to load 'df_pollution') ...

# ==============================================================================
# ### CHANGED: Step 8 - 6-Hour Aggregation (Matching Traffic Data Logic)
# ==============================================================================
# REASON: The previous code took a "snapshot" at specific hours. Pollution is noisy.
# We now take the average of the *passed* 6 hours, matching your traffic data methodology.
# This smoothing helps the AI learn trends rather than random noise.

print("Step 8: Aggregating to 6-hour windows (matching Traffic data logic)...")

# 1. Sort by time
df_pollution = df_pollution.sort_values('timestamp')

# 2. Set index for resampling
df_pollution.set_index('timestamp', inplace=True)

# 3. Group by Station -> Resample to 6H mean
# 'closed=right', 'label=right' ensures that the bin labeled '12:00'
# contains the average of data from 06:00 to 12:00.
pollution_agg = df_pollution.groupby(
    ['pollutant', 'latitude', 'longitude']
)['concentration'].resample('6H', closed='right', label='right').mean().reset_index()

# 4. Cleanup
pollution_agg = pollution_agg.rename(columns={'timestamp': 'time'})
pollution_agg['time'] = pollution_agg['time'].dt.tz_localize(None)

print(f"Step 8: Aggregation complete. {len(pollution_agg)} records.")

# Step 9: Pivot (Standardize format)
pollution_agg = pollution_agg.pivot_table(
    index=['time', 'latitude', 'longitude'],
    columns='pollutant',
    values='concentration'
).reset_index()

# Step 10: Align with ERA5
era5_times = pd.to_datetime(era5_grid['time']).tz_localize(None)
pollution_agg = pollution_agg[pollution_agg['time'].isin(era5_times)]
print(f"Step 10: After exact time alignment → {len(pollution_agg)} rows")

from scipy.spatial import cKDTree  # <--- NEW: Required for Distance Boundary

# ==============================================================================
# ### CORRECTED: Step 11 - Hybrid Interpolation with PHYSICAL BOUNDARY
# ==============================================================================
print("\nStep 11: Starting Hybrid Interpolation with Distance Cutoff...")

times = pd.to_datetime(era5_grid['time'], utc=True).tz_convert(None)
lats = era5_grid['lat']
lons = era5_grid['lon']

# CONFIGURATION: Max distance to project pollution from a station
# 0.5 degrees is approx 55km.
# This prevents "Smearing". Points >55km from a sensor will remain NaN.
MAX_DIST_DEG = 0.5

# 1. Create Target Grid
grid_lat, grid_lon = np.meshgrid(lats, lons, indexing='ij')
grid_points = np.column_stack((grid_lat.ravel(), grid_lon.ravel()))

pollutant_arrays = {p: np.full((len(times), len(lats), len(lons)), np.nan, dtype=float) for p in pollutants.keys()}

for t, time in enumerate(times):
    df_time = pollution_agg[pollution_agg['time'] == time]

    if df_time.empty:
        continue

    # Station coordinates for this hour
    points = df_time[['latitude', 'longitude']].values

    for pollutant in pollutants:
        if pollutant not in df_time: continue

        values = df_time[pollutant].values

        # Valid data only
        mask = ~np.isnan(values)
        valid_points = points[mask]
        valid_values = values[mask]

        if len(valid_values) == 0:
            continue

        try:
            # === STRATEGY: Linear First, Then Radius-Limited Nearest ===

            # 1. Linear Interpolation (Good for smooth gradients between cities)
            # This works inside the "convex hull" of stations
            if len(valid_points) >= 4:
                interp_linear = griddata(
                    valid_points, valid_values, grid_points,
                    method='linear', fill_value=np.nan
                )
            else:
                interp_linear = np.full(len(grid_points), np.nan)

            # 2. Radius-Limited Nearest Neighbor (Fills gaps, but stops at boundary)
            # We use cKDTree to find neighbors within MAX_DIST_DEG
            tree = cKDTree(valid_points)
            dists, idxs = tree.query(grid_points, k=1, distance_upper_bound=MAX_DIST_DEG)

            # 'idxs' will equal len(valid_points) if no neighbor is found within distance
            valid_neighbor_mask = idxs < len(valid_points)

            interp_nearest = np.full(len(grid_points), np.nan)

            # Only fill pixels that actually have a close neighbor
            if np.any(valid_neighbor_mask):
                interp_nearest[valid_neighbor_mask] = valid_values[idxs[valid_neighbor_mask]]

            # 3. Combine: Prefer Linear, fill NaNs with Bounded Nearest
            final_flat = np.where(np.isnan(interp_linear), interp_nearest, interp_linear)

            # Reshape back to (Lat, Lon)
            final_grid = final_flat.reshape(len(lats), len(lons))

            # === PHYSICAL BOUNDS CLIPPING ===
            lower, upper = valid_ranges[pollutant]
            # Warning: Clipping NaNs keeps them as NaN (which is correct for Ocean)
            final_grid = np.clip(final_grid, lower, upper)

            pollutant_arrays[pollutant][t] = final_grid

        except Exception as e:
            print(f"Interpolation error {time} {pollutant}: {e}")

    if t % 100 == 0:
        print(f"Processed {t}/{len(times)} timesteps")

# === STEP 12: Create xarray Dataset (Unchanged) ===
ds = xr.Dataset(
    {
        f'{pollutant}_concentration': (
            ['time', 'lat', 'lon'],
            pollutant_arrays[pollutant],
            {
                'units': 'various', # Ideally fetch from original df
                'long_name': f'{pollutant} concentration',
                'description': 'Hybrid Interpolation (Linear primary, Nearest fallback)',
                'aggregation': 'Mean of passed 6 hours'
            }
        ) for pollutant in pollutants
    },
    coords={
        'time': times,
        'lat': lats,
        'lon': lons
    }
)

ds.to_netcdf(output_file)
print(f"\nSUCCESS: File saved to {output_file}")
```

How 6-Hourly Aggregation Was Performed
The previous script aggregated daily HMS fire data (from hms_fireYYYYMMDD.zip) to 6-hourly intervals to align with ERA5, NOAA, and PeMS (00:00, 06:00, 12:00, 18:00 UTC). Here’s how it was done and why sum (not mean) was used:
Aggregation Process

Input Data:

Each shapefile (e.g., hms_fire20220101.shp) contains fire points with columns like Lon, Lat, YearDay, Time, FRP.
After filtering for LA (lat: 33–35, lon: -119.5 to -116.5), you had 9,766 features across 331 days.
Each feature was assigned fire_presence=1 (binary indicator) and fire_intensity=FRP (MW, or 0 if missing).


Time Assignment:

Daily files were assigned a date (e.g., 2022-01-01 from hms_fire20220101.zip).
Dates were floored to the nearest 6-hour UTC interval using dt.floor('6h'), typically 00:00 UTC of the day, as HMS data lacks sub-daily timestamps in your sample (Time column may not be reliable).


Aggregation:

Grouped by time_6h (6-hourly timestamp).
Applied the following aggregations (from agg_dict):

fire_presence: Sum to count the number of fire points in each 6-hour window (e.g., 25 fires on 2022-01-01 00:00).
fire_intensity: Sum to total the Fire Radiative Power (MW) in each 6-hour window (e.g., 150.5 MW).
latitude, longitude: Mean of centroids to represent the average location of fires (used for interpolation).


Result: 331 records (one per 6-hour period with fires), as most days had no fires in LA.


Interpolation:

Fire points were interpolated to the 9×13 ERA5 grid using griddata with method='nearest'.
Non-fire time steps were filled with zeros, ensuring ~2914 time steps total.



Why Sum Instead of Mean?

Fire Presence:

Sum: Counts the number of fire points in each 6-hour window (e.g., 25 fires → fire_presence=25). This preserves the total number of events, which is critical for sparse, event-based data like fires.
Mean would average the binary indicator (1), resulting in fractional values (e.g., 25/25=1), losing the count of events and making interpretation unclear.


Fire Intensity:

Sum: Totals FRP (MW) across all fire points in a 6-hour window (e.g., 150.5 MW), reflecting the cumulative fire energy, which is meaningful for fire impact.
Mean would average FRP, reducing intensity for multiple fires and making it less representative (e.g., 150.5/25=6.02 MW per fire).


Rationale: Fires are discrete events, and summing preserves the total count and energy, aligning with meteorological applications like GraphCast, where total fire activity matters. Mean is better for continuous variables (e.g., temperature), but not for sparse, count-based data.

Why 6-Hourly?

Matches ERA5, NOAA, and PeMS (6-hourly UTC: 00:00, 06:00, 12:00, 18:00).
Daily HMS data lacks precise sub-daily timestamps, so flooring to 00:00 UTC is a reasonable approximation. If the Time column in shapefiles provides hours, we could refine to specific 6-hour slots.

Notes on Alignment

Units:

fire_presence: Count (unitless, e.g., 0–10).
fire_intensity: MW (e.g., 0–500).
Aligns with ERA5 (t2m, msl_hpa), NOAA (temperature_isd, pressure_isd), and PeMS (traffic metrics).


Grid/Time: 9×13 grid (lat: 33.0–35.0, lon: -119.5 to -116.5), ~2914 time steps (2022–2023).
Sparse Data: Only 331/2914 time steps have fires, so the CSV will be mostly zeros, which is expected.


1. Why Sum Was Used for 6-Hourly Aggregation
The HMS fire data, derived from daily shapefiles (hms_fireYYYYMMDD.zip), was aggregated to 6-hourly intervals to align with ERA5, NOAA, and PeMS datasets (6-hourly UTC: 00:00, 06:00, 12:00, 18:00). The script used sum for fire_presence and fire_intensity during aggregation. Here’s the reasoning:
Aggregation Process Recap

Input: Each shapefile contains fire points with columns like Lon, Lat, YearDay, Time, FRP (Fire Radiative Power in MW). After filtering for the LA region, you had 9,766 features across 331 days.
Variables:

fire_presence: Assigned 1 for each fire point (binary indicator).
fire_intensity: Set to FRP (MW) or 0 if missing.


Aggregation:

Daily data was floored to 6-hour intervals (e.g., 2022-01-01 → 2022-01-01 00:00:00 UTC) using dt.floor('6h').
Grouped by time_6h and aggregated:

fire_presence: Sum to count total fire points in each 6-hour window.
fire_intensity: Sum to total FRP (MW) in each 6-hour window.
latitude, longitude: Mean to estimate the average fire location for interpolation.




Output: 331 6-hourly records (from days with fires), interpolated to the 9×13 ERA5 grid, with zeros for non-fire periods (~2914 total time steps).

Why Sum?

Fire Presence:

Sum: Counts the number of fire points in a 6-hour window (e.g., 25 fires → fire_presence=25). This preserves the total number of fire events, which is critical for sparse, event-based data. It reflects the extent of fire activity, important for GraphCast, which may use fire counts as a proxy for wildfire impact.
Why not mean?: Averaging fire_presence (which is 1 per point) would yield fractional values (e.g., 25/25=1), losing the count of fires and making it hard to interpret the number of events. For example, 25 fires and 1 fire would both yield a mean of 1, which is misleading.
Why not max?: Taking the maximum of fire_presence (always 1) would result in 1 for any period with at least one fire, losing information about multiple fires (e.g., 25 vs. 1 fire).


Fire Intensity:

Sum: Totals FRP across all fire points in a 6-hour window (e.g., 150.5 MW). This captures the cumulative energy released by fires, a key indicator of fire severity, relevant for atmospheric modeling (e.g., aerosol emissions in GraphCast).
Why not mean?: Averaging FRP (e.g., 150.5/25=6.02 MW) would represent the average intensity per fire but lose the total energy, which is more relevant for regional impact. For example, 25 small fires (150.5 MW total) vs. 1 large fire (150.5 MW) would have different atmospheric effects.
Why not max?: Taking the maximum FRP would select only the most intense fire in a 6-hour window (e.g., 10 MW), ignoring other fires’ contributions, underestimating total fire activity.


Rationale: Summing is standard for event-based, additive phenomena like fires, where the total count (fire_presence) and energy (fire_intensity) matter. This aligns with meteorological applications, where cumulative fire activity drives smoke and aerosol impacts. Mean or max would obscure the scale of fire events, especially in sparse data (only 331/730 days with fires).

When to Use Mean or Max?

Mean: Suitable for continuous variables (e.g., temperature, smoke density) to smooth fluctuations. For fires, it dilutes event counts and intensity, making it less appropriate.
Max: Useful for capturing peak events (e.g., maximum wind speed). For fires, it would highlight only the most intense fire, ignoring others, which is less relevant for GraphCast’s regional modeling.
Use Case Check: If GraphCast expects a different aggregation (e.g., max intensity for extreme events), we could modify fire_intensity to use max. However, sum is more consistent with fire monitoring practices (e.g., HMS, MODIS fire products).

```python
from google.colab import drive
drive.mount('/content/drive')

# Install necessary libraries
!pip install xarray pandas numpy geopandas fiona shapely netCDF4

import xarray as xr
import pandas as pd
import numpy as np
import geopandas as gpd
import glob
import zipfile
import os
import gc
import shutil

# --- Configuration ---
hms_dir = '/content/drive/MyDrive/graphcast_project/Fire/'
# Renamed output to reflect the larger domain
output_file = '/content/drive/MyDrive/graphcast_project/Training Data/hms_western_us_fire_2022_2025.nc'
os.makedirs(os.path.dirname(output_file), exist_ok=True)

# 1. Define Expanded Grid (Western US)
# This captures upwind smoke sources (e.g. NorCal, Oregon)
lat_min, lat_max = 28.0, 50.0
lon_min, lon_max = -135.0, -105.0
resolution = 0.25

# Create coordinate arrays (Grid Centers)
lats = np.arange(lat_min, lat_max + resolution, resolution)
lons = np.arange(lon_min, lon_max + resolution, resolution)

# Create TIMEZONE-AWARE index for processing (aligns with HMS data)
# We keep this as UTC during processing to ensure logic is correct
times = pd.date_range('2022-01-01 00:00:00', '2025-11-27 18:00:00', freq='6h', tz='UTC')

print(f"Target Grid: {len(lats)} Lats x {len(lons)} Lons x {len(times)} Times")

# --- Helper Function: Time Parsing ---
def parse_hms_time(df, default_date):
    """Parses HMS shapefile columns (Year, Day, Time) into precise UTC datetimes."""
    try:
        if all(c in df.columns for c in ['Year', 'Day', 'Time']):
            # 1. Base Date from Year + Julian Day
            base_dates = pd.to_datetime(df['Year'].astype(str) + df['Day'].astype(str), format='%Y%j')

            # 2. Parse Time String (HHMM)
            time_str = df['Time'].astype(str).str.split('.').str[0]
            time_str = time_str.str.zfill(4).replace('nan', '0000').replace('0nan', '0000')

            # Extract Hours and Minutes
            hours = pd.to_timedelta(pd.to_numeric(time_str.str[:2], errors='coerce').fillna(0), unit='h')
            mins = pd.to_timedelta(pd.to_numeric(time_str.str[2:], errors='coerce').fillna(0), unit='m')

            return (base_dates + hours + mins).dt.tz_localize('UTC')

        elif 'Date' in df.columns and 'Time' in df.columns:
            return pd.to_datetime(df['Date'].astype(str) + ' ' + df['Time'].astype(str)).dt.tz_localize('UTC')

    except Exception as e:
        pass # Fallback

    # Fallback: Use file date + 12 hours
    return default_date + pd.Timedelta(hours=12)

# --- Main Processing ---

# 1. Setup Histogram Bins
lat_edges = np.append(lats - resolution/2, lats[-1] + resolution/2)
lon_edges = np.append(lons - resolution/2, lons[-1] + resolution/2)

# Initialize Output Arrays
n_times, n_lats, n_lons = len(times), len(lats), len(lons)
grid_count = np.zeros((n_times, n_lats, n_lons), dtype=np.float32)
grid_intensity = np.zeros((n_times, n_lats, n_lons), dtype=np.float32)

# 2. Extract and Filter Data
zip_files = sorted(glob.glob(os.path.join(hms_dir, 'hms_fire*.zip')))
print(f"Step 1: Found {len(zip_files)} HMS zip files")

chunk_list = []
temp_dir = '/tmp/hms_temp_proc/'
os.makedirs(temp_dir, exist_ok=True)

for i, zip_path in enumerate(zip_files):
    try:
        filename = os.path.basename(zip_path)
        date_str = filename.replace('hms_fire', '').replace('.zip', '')
        file_date = pd.to_datetime(date_str, format='%Y%m%d').tz_localize('UTC')

        with zipfile.ZipFile(zip_path, 'r') as z:
            shp_files = [f for f in z.namelist() if f.endswith('.shp')]
            if not shp_files: continue

            z.extractall(temp_dir)

            # Load Shapefile
            gdf = gpd.read_file(os.path.join(temp_dir, shp_files[0]))

            # Spatial Filter
            gdf = gdf.cx[lon_min:lon_max, lat_min:lat_max]

            if not gdf.empty:
                # Parse Time
                gdf['timestamp'] = parse_hms_time(gdf, file_date)

                # Binning: Ceil to 6h
                gdf['time_bin'] = gdf['timestamp'].dt.ceil('6h')

                # Clean FRP
                if 'FRP' in gdf.columns:
                    gdf['FRP'] = pd.to_numeric(gdf['FRP'], errors='coerce').fillna(0).clip(lower=0)
                else:
                    gdf['FRP'] = 0.0

                # Extract Coords
                gdf['lat'] = gdf.geometry.y
                gdf['lon'] = gdf.geometry.x

                chunk_list.append(gdf[['time_bin', 'lat', 'lon', 'FRP']].copy())

            # Cleanup
            for f in os.listdir(temp_dir):
                os.remove(os.path.join(temp_dir, f))

    except Exception as e:
        print(f"Error processing {zip_path}: {e}")
        continue

    if i % 50 == 0:
        print(f"  Processed {i}/{len(zip_files)} files...")
        gc.collect()

shutil.rmtree(temp_dir)

if not chunk_list:
    raise ValueError("No valid fire data found in the specified range.")

print("Step 2: Concatenating all fire points...")
all_fires = pd.concat(chunk_list, ignore_index=True)
del chunk_list
gc.collect()

# Filter to final Time Range
all_fires = all_fires[(all_fires['time_bin'] >= times[0]) & (all_fires['time_bin'] <= times[-1])]

# 3. Binning (Aggregation)
print("Step 3: Aggregating into ERA5 grid...")

for t_idx, t in enumerate(times):
    df_step = all_fires[all_fires['time_bin'] == t]

    if not df_step.empty:
        # Intensity: Sum
        H_int, _, _ = np.histogram2d(
            df_step['lat'], df_step['lon'],
            bins=[lat_edges, lon_edges],
            weights=df_step['FRP']
        )
        grid_intensity[t_idx, :, :] = H_int

        # Presence: Count
        H_count, _, _ = np.histogram2d(
            df_step['lat'], df_step['lon'],
            bins=[lat_edges, lon_edges]
        )
        grid_count[t_idx, :, :] = H_count

    if t_idx % 200 == 0:
        print(f"  Gridded time step {t_idx}/{len(times)}")

# 4. Save to NetCDF (FIXED)
print("Step 4: Saving NetCDF...")

# --- CRITICAL FIX ---
# Convert time index to "naive" UTC before saving.
times_naive = times.tz_convert(None)

ds = xr.Dataset(
    data_vars={
        'fire_presence': (['time', 'lat', 'lon'], grid_count, {'units': 'count'}),
        'fire_intensity': (['time', 'lat', 'lon'], grid_intensity, {'units': 'MW'})
    },
    coords={
        'time': times_naive,  # <--- Use naive times
        'lat': lats,
        'lon': lons
    }
)

ds.attrs['description'] = 'HMS Fire Data (Western US)'
ds.attrs['method'] = 'Sum aggregation, 6h Ceil binning'

ds.to_netcdf(output_file)
print(f"Saved: {output_file}")
```

```python
# Install Rasterio (Standard tool for burning shapes to grids)
!pip install xarray pandas numpy geopandas fiona shapely rasterio netCDF4

import xarray as xr
import pandas as pd
import numpy as np
import geopandas as gpd
import rasterio
from rasterio import features
from rasterio.transform import from_origin
import glob
import zipfile
import os
import gc
import shutil

# -------------------------------
# CONFIGURATION
# -------------------------------
hms_dir = '/content/drive/MyDrive/graphcast_project/Smoke/'
# Naming it 'western_us' to match your wide domain
output_file = '/content/drive/MyDrive/graphcast_project/Training Data/hms_western_us_smoke_2022_2025.nc'

# Ensure output directory exists
os.makedirs(os.path.dirname(output_file), exist_ok=True)

# 1. Define Domain (Matches your Traffic/Pollution Grid)
lat_min, lat_max = 28.0, 50.0
lon_min, lon_max = -135.0, -105.0
grid_res = 0.25

# Generate Coordinates (Rounded to 2 decimals to match Traffic/Pollution precision)
lats = np.round(np.arange(lat_min, lat_max + 0.01, grid_res), 2)
lons = np.round(np.arange(lon_min, lon_max + 0.01, grid_res), 2)
times = pd.date_range('2022-01-01 00:00:00', '2025-11-27 18:00:00', freq='6h', tz='UTC')

print(f"Target Grid: {len(lats)} Lats x {len(lons)} Lons x {len(times)} Times")

# -------------------------------
# 2. Setup Data Mappings
# -------------------------------
# Map Smoke Strings to Numeric Density (0-1 Scale)
# Light=0.1, Medium=0.5, Heavy=1.0
DENSITY_MAP = {
    'Light': 0.1, 'light': 0.1, 'LIGHT': 0.1, 'L': 0.1,
    'Medium': 0.5, 'medium': 0.5, 'MEDIUM': 0.5, 'M': 0.5,
    'Heavy': 1.0, 'heavy': 1.0, 'HEAVY': 1.0, 'H': 1.0
}

def parse_hms_time(julian_str, default_date):
    """Parses HMS 'YYYYDDD HHMM' format."""
    try:
        parts = str(julian_str).strip().split()
        if len(parts) != 2: return pd.NaT
        year_day, time_str = parts

        # Parse Year/Day/Time
        year = int(year_day[:4])
        day = int(year_day[4:])
        hour = int(time_str[:2])
        minute = int(time_str[2:])

        dt = pd.to_datetime(f'{year}-01-01') + pd.Timedelta(days=day-1)
        return dt.replace(hour=hour, minute=minute).tz_localize('UTC')
    except:
        return default_date

# -------------------------------
# 3. Load & Process Shapefiles
# -------------------------------
zip_files = sorted(glob.glob(os.path.join(hms_dir, 'hms_smoke*.zip')))
print(f"Found {len(zip_files)} zip files.")

temp_dir = '/tmp/hms_smoke_temp/'
os.makedirs(temp_dir, exist_ok=True)

gdf_list = []

print("Step 1: extracting and parsing shapefiles...")
for i, zip_path in enumerate(zip_files):
    try:
        # Get fallback date from filename
        filename = os.path.basename(zip_path)
        date_str = filename.replace('hms_smoke', '').replace('.zip', '')
        file_date = pd.to_datetime(date_str, format='%Y%m%d').tz_localize('UTC')

        with zipfile.ZipFile(zip_path, 'r') as z:
            shp_files = [f for f in z.namelist() if f.endswith('.shp')]
            if not shp_files: continue

            z.extractall(temp_dir)
            gdf = gpd.read_file(os.path.join(temp_dir, shp_files[0]))

            # Spatial Filter (Broad Box)
            gdf = gdf.cx[lon_min:lon_max, lat_min:lat_max]

            if not gdf.empty:
                # 1. Parse Times
                gdf['start_t'] = gdf['Start'].apply(lambda x: parse_hms_time(x, file_date))
                gdf['end_t'] = gdf['End'].apply(lambda x: parse_hms_time(x, file_date + pd.Timedelta(hours=23)))

                # 2. Map Density
                gdf['val'] = gdf['Density'].map(DENSITY_MAP).fillna(0.1)

                # Keep lightweight columns
                gdf_list.append(gdf[['geometry', 'start_t', 'end_t', 'val']].copy())

            # Cleanup
            for f in os.listdir(temp_dir): os.remove(os.path.join(temp_dir, f))

    except Exception as e:
        print(f"Error {zip_path}: {e}")
        continue

    if i % 20 == 0:
        print(f"  Processed {i}/{len(zip_files)} files...")
        gc.collect()

shutil.rmtree(temp_dir)

if not gdf_list:
    raise ValueError("No valid smoke data found!")

print("Step 2: Combining polygons...")
all_smoke = pd.concat(gdf_list, ignore_index=True)
del gdf_list
gc.collect()

# -------------------------------
# 4. Rasterization (Burning to Grid)
# -------------------------------
print("Step 3: Rasterizing polygons to grid (Max Density logic)...")

# Initialize Output Array [Time, Lat, Lon]
smoke_grid = np.zeros((len(times), len(lats), len(lons)), dtype=np.float32)

# Define Transform: Maps Array Indices (Row, Col) to Lat/Lon
# We use a custom Affine transform to ensure "South-Up" orientation matches your Arrays
# Pixel(0,0) Center will match exactly (lon_min, lat_min)
custom_transform = rasterio.Affine(
    grid_res, 0.0, lon_min - grid_res/2,
    0.0, grid_res, lat_min - grid_res/2
)

# Iterate through TIME
for t_idx, t in enumerate(times):

    # 1. Filter: Find polygons active in this 6H window
    # Logic: Plume must exist within [T-6h, T]
    # This matches the "Ceil" binning (accumulating the past 6 hours)
    window_start = t - pd.Timedelta(hours=6)

    current_polys = all_smoke[
        (all_smoke['start_t'] <= t) &
        (all_smoke['end_t'] >= window_start)
    ]

    if not current_polys.empty:
        # 2. Sort by Density
        # Rasterio "burns" sequentially. By sorting Light -> Heavy,
        # Heavy will overwrite Light if they overlap. This equals MAX aggregation.
        current_polys = current_polys.sort_values('val', ascending=True)

        shapes = ((geom, val) for geom, val in zip(current_polys.geometry, current_polys.val))

        try:
            # Burn
            burned = features.rasterize(
                shapes=shapes,
                out_shape=(len(lats), len(lons)),
                transform=custom_transform,
                fill=0,
                all_touched=True, # Important: Capture smoke even if it just touches the pixel
                dtype=np.float32
            )

            # Save to array
            smoke_grid[t_idx, :, :] = burned

        except Exception as e:
            print(f"Rasterize error at {t}: {e}")

    if t_idx % 100 == 0:
        print(f"  Rasterized {t_idx}/{len(times)} steps")

# -------------------------------
# 5. Save NetCDF
# -------------------------------
print("Step 4: Saving NetCDF...")

times_naive = times.tz_convert(None)

ds = xr.Dataset(
    data_vars={
        'smoke_density': (
            ['time', 'lat', 'lon'],
            smoke_grid,
            {
                'units': 'index (0-1)',
                'description': 'HMS Smoke Density (0.1=Light, 0.5=Med, 1.0=Heavy), Max Aggregation'
            }
        )
    },
    coords={
        'time': times_naive,
        'lat': lats,
        'lon': lons
    }
)

ds.attrs['description'] = 'HMS Smoke Polygons rasterized to 0.25deg Grid (Western US)'
ds.attrs['method'] = 'Rasterization (All Touched), Max Density for overlaps'

encoding = {
    'smoke_density': {'zlib': True, 'complevel': 5, '_FillValue': -1}
}

ds.to_netcdf(output_file, encoding=encoding)
print(f"Saved: {output_file}")
```

```python
from google.colab import drive
drive.mount('/content/drive')

!pip -q install xarray netCDF4 dask[complete] pandas numpy

import os
import numpy as np
import pandas as pd
import xarray as xr

data_dir = '/content/drive/MyDrive/graphcast_project/Training Data/'
files = {
    'ERA5': 'era5_california_2022_2025_WIDE_FULL.nc',
    'PeMS': 'traffic_california_2022_2025_utc_binned_gemini_with_boundary.nc',
    'HMS_Fire': 'hms_western_us_fire_2022_2025.nc',
    'HMS_Smoke': 'hms_western_us_smoke_2022_2025.nc',
    'Pollution': 'pollution_la_2022_to_2025_avg_gemini.nc'
}

# Project expectation (broad)
EXPECTED_START = pd.Timestamp("2022-01-01 00:00:00", tz="UTC")
EXPECTED_END   = pd.Timestamp("2025-05-30 18:00:00", tz="UTC")
EXPECTED_FREQ  = pd.Timedelta(hours=6)
EXPECTED_HOURS = {0, 6, 12, 18}

def _pick_coord_name(ds, candidates):
    for c in candidates:
        if c in ds.coords:
            return c
        if c in ds.variables:
            return c
    return None

def _to_datetime_index(time_values):
    """
    Robustly convert xarray time coord to pandas.DatetimeIndex (UTC-naive).
    Handles numpy datetime64, pandas timestamps, and cftime objects.
    """
    try:
        t = pd.to_datetime(time_values)
        # make timezone-naive but in UTC interpretation
        if getattr(t, "tz", None) is not None:
            t = t.tz_convert("UTC").tz_localize(None)
        return pd.DatetimeIndex(t)
    except Exception:
        # fallback: string conversion
        t = pd.to_datetime([str(x) for x in time_values])
        return pd.DatetimeIndex(t)

def _time_cadence_report(t: pd.DatetimeIndex):
    t = pd.DatetimeIndex(t).sort_values()
    diffs = pd.Series(t[1:] - t[:-1])
    mode = diffs.mode().iloc[0] if len(diffs) else None
    irregular = int((diffs != EXPECTED_FREQ).sum()) if len(diffs) else 0
    dup = int(t.duplicated().sum())
    hours = set(pd.Series(t.hour).unique())
    return {
        "start": t.min(),
        "end": t.max(),
        "n": len(t),
        "dup": dup,
        "mode_step": mode,
        "irregular_steps": irregular,
        "hours_present": hours,
    }

def validate_dataset(ds, name, chunks_note=True, sample_time=6):
    print(f"\n==================== {name} ====================")
    print("sizes:", dict(ds.sizes))
    print("data_vars:", list(ds.data_vars))

    # Detect coord names
    time_name = _pick_coord_name(ds, ["time"])
    lat_name  = _pick_coord_name(ds, ["lat", "latitude", "y"])
    lon_name  = _pick_coord_name(ds, ["lon", "longitude", "x"])

    print("coord names:", {"time": time_name, "lat": lat_name, "lon": lon_name})
    if time_name is None:
        print("⚠️  No time coordinate found. Skipping time checks.")
        return None, (lat_name, lon_name)

    # Time checks
    t = _to_datetime_index(ds[time_name].values)
    rep = _time_cadence_report(t)
    print("time:", rep["start"], "→", rep["end"], "| n=", rep["n"])
    print("duplicates:", rep["dup"], "| mode step:", rep["mode_step"], "| irregular steps:", rep["irregular_steps"])
    print("hours present:", sorted(list(rep["hours_present"])))

    # Compare to expectation in a tolerant way (don’t require exact equality)
    if rep["start"] != EXPECTED_START.tz_localize(None) or rep["end"] != EXPECTED_END.tz_localize(None):
        print(f"⚠️  Time range differs from expected {EXPECTED_START} → {EXPECTED_END} (UTC).")
    if rep["mode_step"] is not None and rep["mode_step"] != EXPECTED_FREQ:
        print(f"⚠️  Expected 6-hour cadence but mode step is {rep['mode_step']}")
    if not (rep["hours_present"] <= EXPECTED_HOURS):
        print(f"⚠️  Unexpected hours present. Expected subset of {sorted(EXPECTED_HOURS)}")

    # Lat/lon checks (range + monotonic)
    if lat_name is not None:
        lat = ds[lat_name].values
        print("lat range:", float(np.nanmin(lat)), "→", float(np.nanmax(lat)), "len=", len(lat),
              "| monotonic:", bool(np.all(np.diff(lat) != 0)))
    if lon_name is not None:
        lon = ds[lon_name].values
        print("lon range:", float(np.nanmin(lon)), "→", float(np.nanmax(lon)), "len=", len(lon),
              "| monotonic:", bool(np.all(np.diff(lon) != 0)))

    # Variable-wise missingness and quick stats WITHOUT loading everything
    # We'll sample a few time indices to estimate distributions.
    if "time" in ds.dims:
        nT = ds.sizes["time"]
        sample_idx = np.linspace(0, nT-1, min(sample_time, nT)).astype(int)
    else:
        sample_idx = None

    for v in ds.data_vars:
        da = ds[v]
        units = da.attrs.get("units", "unknown")
        print(f"\n- {v} (units: {units}) dims={da.dims} dtype={da.dtype}")

        # NaN percent computed lazily if dask-backed
        try:
            nan_frac = da.isnull().mean().compute().item() if hasattr(da.data, "compute") else float(da.isnull().mean().values)
            print(f"  NaN fraction: {nan_frac*100:.3f}%")
        except Exception as e:
            print(f"  (NaN fraction calc failed: {e})")

        # Quick percentiles from sampled time slices to avoid full read
        try:
            sample = da
            if sample_idx is not None and "time" in da.dims:
                sample = da.isel(time=sample_idx)
            # compute small sample fully
            x = sample.values
            if np.isfinite(x).any():
                q = np.nanpercentile(x, [0, 1, 5, 50, 95, 99, 100])
                print("  pct [0,1,5,50,95,99,100]:", np.array2string(q, precision=4))
            else:
                print("  all missing in sample.")
        except Exception as e:
            print(f"  (percentile sample failed: {e})")

    return t, (lat_name, lon_name)

# Open datasets with chunking so validation doesn't read everything
datasets = {}
times_dict = {}
coord_names = {}

for name, fname in files.items():
    path = os.path.join(data_dir, fname)
    if not os.path.exists(path):
        print("❌ Missing:", path)
        continue

    print("\nOpening:", name, "->", path)
    ds = xr.open_dataset(path, chunks={"time": 64})  # adjust chunk size if needed
    datasets[name] = ds
    t, (lat_name, lon_name) = validate_dataset(ds, name)
    times_dict[name] = t
    coord_names[name] = (lat_name, lon_name)

# Cross-dataset time alignment check (robust: intersection size & mismatches)
print("\n==================== Time Alignment Summary ====================")
ref_name = "ERA5" if "ERA5" in times_dict else list(times_dict.keys())[0]
ref_t = times_dict.get(ref_name)

if ref_t is None:
    print("No reference time axis available.")
else:
    ref_set = set(ref_t)
    for name, t in times_dict.items():
        if t is None:
            continue
        s = set(t)
        inter = len(ref_set & s)
        print(f"{name} vs {ref_name}:")
        print("  n_ref:", len(ref_t), "n_this:", len(t), "intersection:", inter)
        if inter < min(len(ref_t), len(t)):
            missing_in_this = sorted(list(ref_set - s))[:5]
            extra_in_this = sorted(list(s - ref_set))[:5]
            print("  ⚠️ mismatch examples:")
            if missing_in_this: print("    missing in this (first 5):", missing_in_this)
            if extra_in_this:   print("    extra in this (first 5):", extra_in_this)

# Close when done
for ds in datasets.values():
    ds.close()
```

Step 10)combine all files into a single file.

Data Massaging (Preprocessing)
Run this code in Colab. It creates a single .npz file containing the aligned Inputs, Targets, and Validity Mask.

Key Logic:

It uses the Pollution grid as the "Anchor" and snaps ERA5/Traffic to it.

It creates a Mask where 1 = Real Data and 0 = Ocean/Missing.

```python
import xarray as xr
import numpy as np
import pandas as pd
import os

# ==========================================
# CONFIGURATION
# ==========================================
DATA_DIR = '/content/drive/MyDrive/graphcast_project/Training Data/'
SAVE_PATH = os.path.join(DATA_DIR, 'processed_simvp_data.npz')

FILES = {
    'ERA5': 'era5_california_2022_2025_WIDE_FULL.nc',
    'PeMS': 'traffic_california_2022_2025_utc_binned_gemini_with_boundary.nc',
    'Fire': 'hms_western_us_fire_2022_2025.nc',
    'Smoke': 'hms_western_us_smoke_2022_2025.nc',
    'Pollution': 'pollution_la_2022_to_2025_avg_gemini.nc'
}

def preprocess_pipeline():
    print(">>> 1. Loading Datasets...")
    dsets = {k: xr.open_dataset(os.path.join(DATA_DIR, v)) for k, v in FILES.items()}

    # ==========================================
    # A. SPATIAL ALIGNMENT (Crop to 88x120)
    # ==========================================
    print(">>> 2. Aligning Grids...")
    # Use Pollution as the Master Grid
    target_lat = dsets['Pollution'].lat
    target_lon = dsets['Pollution'].lon

    # Snap other datasets to this grid (Nearest Neighbor handles the crop)
    for name in ['ERA5', 'PeMS', 'Fire', 'Smoke']:
        dsets[name] = dsets[name].sel(lat=target_lat, lon=target_lon, method='nearest')

    print(f"    Grid Shape Aligned to: {len(target_lat)} x {len(target_lon)}")

    # ==========================================
    # B. TEMPORAL ALIGNMENT (Intersection)
    # ==========================================
    print(">>> 3. Aligning Time...")
    # Find timestamps that exist in ERA5 and Pollution (most critical)
    t1 = dsets['ERA5'].time.values
    t2 = dsets['Pollution'].time.values
    common_times = np.intersect1d(t1, t2)

    print(f"    Common Time Steps: {len(common_times)}")

    # Reindex all datasets to these common times
    # 'ffill' fixes the missing gaps in PeMS
    for k in dsets:
        dsets[k] = dsets[k].reindex(time=common_times, method='ffill')

    # ==========================================
    # C. NORMALIZATION & MASKING
    # ==========================================
    print(">>> 4. Normalizing & Stacking...")
    input_list = []

    # 1. WEATHER (Z-Score)
    # --------------------
    print("    Processing Weather (ERA5)...")
    # Convert to (Time, Channels, H, W)
    era_data = dsets['ERA5'].to_array().values.transpose(1, 0, 2, 3)
    mean = np.nanmean(era_data, axis=(0, 2, 3), keepdims=True)
    std = np.nanstd(era_data, axis=(0, 2, 3), keepdims=True)
    era_norm = (era_data - mean) / (std + 1e-6)
    input_list.append(np.nan_to_num(era_norm)) # Safety fill

    # 2. SPARSE DATA (Log Transform)
    # ------------------------------
    print("    Processing Traffic/Fire/Smoke...")
    for name in ['PeMS', 'Fire', 'Smoke']:
        data = dsets[name].to_array().values.transpose(1, 0, 2, 3)
        data = np.nan_to_num(data, nan=0.0)
        data = np.log1p(data) # Log(x+1) to squash spikes
        # MinMax Scale roughly to 0-1
        mx = np.max(data)
        if mx > 0: data = data / mx
        input_list.append(data)

    # 3. POLLUTION (Target & History)
    # -------------------------------
    print("    Processing Pollution...")
    # Explicit order to ensure we know which channel is which
    pol_vars = ['O3_concentration', 'SO2_concentration', 'CO_concentration',
                'NO2_concentration', 'PM10_concentration', 'PM25_concentration']

    ds_pol = dsets['Pollution'][pol_vars]
    pol_raw = ds_pol.to_array().values.transpose(1, 0, 2, 3)

    # create MASK: 1 if sensor data exists, 0 if Ocean/NaN
    # We use a shared mask based on valid data presence
    valid_mask = (~np.isnan(pol_raw)).astype(np.float32)
    # Reduce to (Time, 1, H, W) if you want a shared mask,
    # OR keep (Time, 6, H, W) for variable-specific masking.
    # Let's use variable-specific to be precise.

    # Fill NaN with 0 for Input
    pol_filled = np.nan_to_num(pol_raw, nan=0.0)
    pol_log = np.log1p(pol_filled)
    pol_max = np.max(pol_log) # Save this if you need to un-normalize later
    pol_norm = pol_log / (pol_max + 1e-6)

    # Add pollution history to inputs
    input_list.append(pol_norm)

    # ==========================================
    # D. SAVE
    # ==========================================
    # X: [ERA5, Traffic, Fire, Smoke, Pollution]
    X = np.concatenate(input_list, axis=1).astype(np.float32)
    Y = pol_norm.astype(np.float32)
    Mask = valid_mask.astype(np.float32)


    print(f"    Final Input Shape: {X.shape}")
    print(f"    Final Target Shape: {Y.shape}")
    print(f"    Saving to {SAVE_PATH}...")
    np.savez_compressed(SAVE_PATH, X=X, Y=Y, Mask=Mask)
    print("✅ Preprocessing Complete!")

preprocess_pipeline()
```
